```python
data=pd.read_csv('countries.csv')
```

```python
data
```

```python
#compare population growth in US and China
```

```python
ind=data[data.country=='India']
ind
```

```python
us=data[data.country=='United States']
us
```

```python
china=data[data.country=='China']
china
```

```python
plt.plot(us.year,us.population / 10**6)
plt.plot(china.year,china.population / 10**6)
plt.title('POPULATION RATE OF US vs CHINA')
plt.legend(['United States','China'])
plt.xlabel("year")
plt.ylabel("population in million")
plt.show
```

```python
us.population
```

```python
us.population/us.population.iloc[0] *100
```

```python
plt.plot(us.year,us.population/us.population.iloc[0] *100)
plt.plot(china.year,china.population/china.population.iloc[0] *100)
plt.plot(ind.year,ind.population/ind.population.iloc[0] *100)
plt.title('RELATIVE POPULATION RATE OF US vs CHINA')
plt.legend(['United States','China','India'])
plt.xlabel("year")
plt.ylabel("population growth (first==100)")
plt.show
```
